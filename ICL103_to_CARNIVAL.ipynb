## Take log2 of gene intensities

```python
log2_icl103_gene_intensities = np.log2(icl103_gene_intensities)
log2_icl103_gene_intensities
```

```python
def explode_column(df, varname, newcols, sep):
    df[newcols] = df[varname].str.split(sep,expand=True)
    return df
```

## Convert to Tidy format

```python
design = ['Study','Strain','Time','SampleNum']
melted_icl103_gene_intensities = log2_icl103_gene_intensities.reset_index().melt(id_vars=gene_columns,
                                                                            var_name='Sample',
                                                                            value_name='Intensity')
melted_icl103_gene_intensities[design] = melted_icl103_gene_intensities['Sample'].str.split('_',expand=True) 
melted_icl103_gene_intensities
```

## Calculate mean and standard error across replicates for each strain and time point

```python
mean_sem_gene_intensities = melted_icl103_gene_intensities.groupby(gene_columns + ['Study','Strain','Time']).agg( ['mean','sem'])
mean_sem_gene_intensities
```

## Calculate $\log_2$ fold change for VN1203 strain vs Mock

```python
log2fc_gene_intensities = mean_sem_gene_intensities.unstack('Strain')['Intensity','mean','VN1203'].subtract(mean_sem_gene_intensities.unstack('Strain')['Intensity','mean','Mock']).to_frame('log2fc')
```

## Combine standard error for VN1203 vs Mock

Combined standard error for VN1203 and Mock is calculated as $\sqrt{SE_{VN1203}^2 + SE_{Mock}^2}$

```python
log2fc_gene_intensities['sem'] = np.sqrt( np.square(mean_sem_gene_intensities.unstack('Strain')['Intensity','sem','VN1203'])+ 
                                          np.square(mean_sem_gene_intensities.unstack('Strain')['Intensity','sem','Mock'] ))
```

## Calculate Wald statistic for input to DOROTHEA

```python
log2fc_gene_intensities['stat'] = log2fc_gene_intensities['log2fc'].divide(log2fc_gene_intensities['sem'])
log2fc_gene_intensities
```

## Calculate $\log_2$ fold change for VN1203 strain vs Mock

```python
log2fc_prot = (mean_sem_prot.unstack('Strain')['Intensity','mean','VN1203']
 .subtract(mean_sem_prot.unstack('Strain')['Intensity','mean','Mock'])
 .to_frame('log2fc'))
log2fc_prot
```

### Combine standard error for VN1203 vs Mock

Combined standard error for VN1203 and Mock is calculated as $\sqrt{SE_{VN1203}^2 + SE_{Mock}^2}$

```python
log2fc_prot['sem'] = np.sqrt( np.square(mean_sem_prot.unstack('Strain')['Intensity','sem','VN1203'])+ 
                              np.square(mean_sem_prot.unstack('Strain')['Intensity','sem','Mock'] ))
log2fc_prot
```

### Calculate Wald statistic for input to DOROTHEA



```python
log2fc_prot['stat'] = log2fc_prot['log2fc'].divide(log2fc_prot['sem'])
swapped_log2fc_prot = log2fc_prot.unstack(0).swaplevel(i=0,j=1,axis=1)
swapped_log2fc_prot
```

```python
vn1203_v_mock_ns1_ko = {}
true_inactivated = {}
false_inactivated = {}
true_derepressed = {}
false_derepressed = {}
predicted_ns1_v_mock_tf = {}
rdir = 'Results'
confusion = {}
experiment = 'VN1203vsMock_for_{}_kegg_influenza_ns1'
for time in ['0hr', '12hr','18hr','24hr']:
    vn1203_v_mock_ns1_ko[time] = pd.read_excel(os.path.join(rdir,
                                                      experiment.format(time),
                                                     experiment.format(time) + '.xlsx'),
                                              sheet_name='Protein Activity',
                                              index_col=0)
    observed = vn1203_v_mock_ns1_ko[time]['nodetype'] == 'FootprintObserved'
    predicted_ns1_v_mock_tf[time] = vn1203_v_mock_ns1_ko[time][observed].join(vn1203_v_mock_tf[time[:-1]], how='left')
    predicted_inactivated = predicted_ns1_v_mock_tf[time]['stat'] > 1.65
    predicted_derepressed = predicted_ns1_v_mock_tf[time]['stat'] < -1.65
    actual_inactivated = ns1_v_mock_tf[time[:-1]]['stat'] < 1.65
    actual_derepressed = ns1_v_mock_tf[time[:-1]]['stat'] > -1.65
    true_inactivated[time] = (set(predicted_ns1_v_mock_tf[time][predicted_inactivated].index) & 
                              set(ns1_v_mock_tf[time[:-1]][actual_inactivated].index))
    false_inactivated[time] = (set(predicted_ns1_v_mock_tf[time][predicted_inactivated].index) & 
                          set(ns1_v_mock_tf[time[:-1]][~actual_inactivated].index))
    true_derepressed[time] = (set(predicted_ns1_v_mock_tf[time][predicted_derepressed].index) & 
                              set(ns1_v_mock_tf[time[:-1]][actual_derepressed].index))
    false_derepressed[time] = (set(predicted_ns1_v_mock_tf[time][predicted_derepressed].index) & 
                          set(ns1_v_mock_tf[time[:-1]][~actual_derepressed].index))
    confusion[time] = pd.DataFrame({'Predicted inactivated': [len(true_inactivated[time]),
                                                       len(false_inactivated[time])],
                             'Predicted derepressed': [len(true_derepressed[time]),
                                                       len(false_derepressed[time])]},
                            index=['Actual inactivated', 'Actual derepressed'])
    print(time)
    display(confusion[time])
```

```python
from pycm import ConfusionMatrix, Compare
def get_confusion_matrix(predicted, actual, significance_threshold=1.65):
    predicted_inactivated = predicted['stat'] > significance_threshold
    predicted_derepressed = predicted['stat'] < -significance_threshold
    actual_inactivated = actual['stat'] < significance_threshold
    actual_derepressed = actual['stat'] > -significance_threshold
    true_inactivated = (set(predicted[predicted_inactivated].index) & 
                        set(actual[actual_inactivated].index))
    false_inactivated = (set(predicted[predicted_inactivated].index) & 
                         set(actual[~actual_inactivated].index))
    true_derepressed = (set(predicted[predicted_derepressed].index) & 
                        set(actual[actual_derepressed].index))
    false_derepressed = (set(predicted[predicted_derepressed].index) & 
                         set(actual[~actual_derepressed].index))
    y_pred = ['inactivated']*len(true_inactivated) + ['inactivated']*len(false_inactivated) + ['derepressed']*len(true_derepressed) + ['derepressed']*len(false_derepressed)
    y_true  = ['inactivated']*len(true_inactivated) + ['derepressed']*len(false_inactivated) + ['derepressed']*len(true_derepressed) + ['inactivated']*len(false_derepressed)
    return ConfusionMatrix(y_true, y_pred), true_inactivated, false_inactivated, true_derepressed,false_derepressed
```
