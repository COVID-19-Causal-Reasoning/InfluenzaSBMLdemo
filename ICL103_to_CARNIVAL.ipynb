# Host virus interactome

CARNIVAL expects a perturbation file, and we can provide 1093 host-virus interactions that were integrated from VirusMentha,Virus MINT, IntAct, and Pfam by [García-Pérez 2017](https://sciwheel.com/work/#/items/9447021/detail?collection=380221&tags=254725), or  136 protein interactions that appeared in 3 out of 4 affinity purification/Mass Spec screens [Zhao 2017](https://sciwheel.com/work/#/items/4447651/detail?collection=380221&tags=254725) (136).  CARNIVAL prefers a smaller number, so we will go with Zhao.

```python
import pandas as pd
zhao = pd.read_excel('zhao_consensus_interactome.xlsx', sheet_name=0, skiprows=2, keep_default_na=False )
zhao.columns
```

```python
zhao['relationship'] = -1
zhao.to_csv('zhao_consensus_interactions.csv', 
            columns=['IAV Protein', 'relationship','Interacting Host Factor'],
            index=False, header=['from','interaction' ,'to'])
with open('zhao_virus_proteins.csv','w') as out:
    virus_proteins = sorted(set(zhao['IAV Protein'].values))
    out.write(',Virusproteins\n')
    out.write('\n'.join([f'{v},1' for v in virus_proteins]))
                            
!cat zhao_consensus_interactions.csv
!cat zhao_virus_proteins.csv
```

## Roll up probe intensities to Gene intensities 

**Note** gene symbols are duplicated across splice variants. Refseq allows unique identifiers for splice variants, but OmniPath just uses HGNC gene symbol. So we will use Gene Symbols and average across splice variants for lack of a better thing to do.

```python
gene_symbols = icl103_gene_intensities.droplevel(['GeneName','Description','RefSeq','EntrezID','GeneSymbol','Ensembl gene ID'])
gene_symbols.loc['HAAO']
```

```python
icl103_gene_intensities = icl103_probe_intensities.groupby(gene_columns).mean()
icl103_gene_intensities
```

## Export normalized gene intensities

Note that these are microarray fluorescence values, which may not have the dynamic range of RNA read counts.
They are also normalized across arrays, not library size, as in RNA-Seq.
Lastly, they are floating values, not integers, as in RNA-Seq (although after normalization, RNA-Seq counts are floats).

Since we could not resolve the Gene Symbol problem for input to Progeny, we kept the row associated with first Gene Symbol, and dropped the rest.  **TODO**  come up with a more elegant way of dealing with this.

```python
for time in ['0hr','7hr','12hr', '24hr']:
    columns = [col for col in icl103_gene_intensities.columns
        if (('Mock' in col) or ('VN1203' in col)) and time in col]
    deduped_gene_intensities = (icl103_gene_intensities[columns]
                                .droplevel(['GeneName','Description','RefSeq','EntrezID', 'GeneSymbol','Ensembl gene ID'])
                                .reset_index()
                                .drop_duplicates('Approved symbol')
                                .set_index('Approved symbol'))
    deduped_gene_intensities.to_csv(f'ICL103_gene_intensities_VN1203_vs_Mock_at_{time}.csv',
                                    index_label=False)
    print(columns)
    display(deduped_gene_intensities)
```

## Take log2 of gene intensities

```python
log2_icl103_gene_intensities = np.log2(icl103_gene_intensities)
log2_icl103_gene_intensities
```

```python
def explode_column(df, varname, newcols, sep):
    df[newcols] = df[varname].str.split(sep,expand=True)
    return df
```

## Convert to Tidy format

```python
design = ['Study','Strain','Time','SampleNum']
melted_icl103_gene_intensities = log2_icl103_gene_intensities.reset_index().melt(id_vars=gene_columns,
                                                                            var_name='Sample',
                                                                            value_name='Intensity')
melted_icl103_gene_intensities[design] = melted_icl103_gene_intensities['Sample'].str.split('_',expand=True) 
melted_icl103_gene_intensities
```

## Calculate mean and standard error across replicates for each strain and time point

```python
mean_sem_gene_intensities = melted_icl103_gene_intensities.groupby(gene_columns + ['Study','Strain','Time']).agg( ['mean','sem'])
mean_sem_gene_intensities
```

## Calculate $\log_2$ fold change for VN1203 strain vs Mock

```python
log2fc_gene_intensities = mean_sem_gene_intensities.unstack('Strain')['Intensity','mean','VN1203'].subtract(mean_sem_gene_intensities.unstack('Strain')['Intensity','mean','Mock']).to_frame('log2fc')
```

## Combine standard error for VN1203 vs Mock

Combined standard error for VN1203 and Mock is calculated as $\sqrt{SE_{VN1203}^2 + SE_{Mock}^2}$

```python
log2fc_gene_intensities['sem'] = np.sqrt( np.square(mean_sem_gene_intensities.unstack('Strain')['Intensity','sem','VN1203'])+ 
                                          np.square(mean_sem_gene_intensities.unstack('Strain')['Intensity','sem','Mock'] ))
```

## Calculate Wald statistic for input to DOROTHEA

```python
log2fc_gene_intensities['stat'] = log2fc_gene_intensities['log2fc'].divide(log2fc_gene_intensities['sem'])
log2fc_gene_intensities
```
